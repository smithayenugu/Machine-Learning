### **Cell 9**


```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

- **Explanation**:  
  - Imports libraries for data visualization (`seaborn`, `matplotlib`) and clustering (`KMeans` from `sklearn`).

---

### **Cell 10**


```python
k_range = range(1, 10)
sse = []
for k in k_range:
    km = KMeans(n_clusters=k, n_init=10)
    km.fit_predict(df[['petalLength', 'petalWidth', 'sepalLength', 'sepalWidth']])
    sse.append(km.inertia_)
```

- **Explanation**:  
  - Implements the Elbow Method to determine the optimal number of clusters.
  - `k_range` defines the range of cluster numbers to test.
  - `sse` stores the sum of squared errors (inertia) for each value of `k`.

---

### **Cell 11**


```python
sse
```

- **Explanation**:  
  - Displays the list of SSE values for each number of clusters.

---

### **Cell 12**


```python
plt.plot(k_range, sse)
plt.show()
```

- **Explanation**:  
  - Plots the Elbow Curve to visualize the optimal number of clusters.
  - The "elbow point" indicates the best value for `k`.

---

### **Cell 13**


```python
km = KMeans(n_clusters=3, n_init=10)
km.fit_predict(df[['petalLength', 'petalWidth', 'sepalLength', 'sepalWidth']])
```

- **Explanation**:
